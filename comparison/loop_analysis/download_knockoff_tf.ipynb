```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
import pandas as pd
import random, time
```

# Download files

```python
driver = webdriver.Chrome()
driver.get(r"http://www.licpathway.net:8081/KnockTFv2/download.php")
time.sleep(random.uniform(0.5, 1))


# type K562
textbox = driver.find_element(By.TAG_NAME, "input")
textbox.clear()
textbox.send_keys("K562")
print("Input typed:", textbox.get_attribute("value"))

time.sleep(random.uniform(0.5, 1))


#  show 100 entries
dropdown = Select(driver.find_element(By.TAG_NAME, "select"))
dropdown.select_by_visible_text("100")
time.sleep(random.uniform(0.5, 1))


links = []
texts = []
page_number = 1
while 1:
    elements = driver.find_elements(By.CSS_SELECTOR, "tr:has(i.fa.fa-download)")
    print(len(elements))
    page_number += 1
    for element in elements:
        link = element.find_element(By.CSS_SELECTOR, "a:has(i.fa.fa-download)")
        link_text = link.get_attribute("href")
        text = element.text
        links += [link_text]
        texts += [text]
        link.click()
        time.sleep(random.uniform(0.5, 2))

    try: 
        driver.find_element(By.CSS_SELECTOR, f'a[aria-controls="download_table"][data-dt-idx="{page_number}"]').click()
        time.sleep(random.uniform(0.5, 2))
    except:
        break



```

```python
df = pd.DataFrame({'Link': links, 'text': texts})
df['TF'] = df.text.apply(lambda x: x.split(' ')[1])
df['name'] = df.Link.apply(lambda x: x.split(r"/")[-1])
df["name"] = df["name"].str.replace(
    r"^DataSet_(\d+)_(\d+).txt$",
    lambda m: "DataSet_" + m.group(1) + '_' +  m.group(2).zfill(3) + '.txt',
    regex=True
)

df.to_csv('save_knockdown.csv', index=False)
```
